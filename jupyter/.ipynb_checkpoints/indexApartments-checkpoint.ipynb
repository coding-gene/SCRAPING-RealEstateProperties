```python
import pandas as pd
```

```python
df = pd.read_csv(r'index_apartments.csv')
df.head(5)
```

```python
df.info()
```

```python
import seaborn as sns
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(11.7,8.27)})
import matplotlib.pylab as plt
```

```python
ax = sns.barplot(x=df['Location'], y=df['PriceEuro'])
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

```python
sns.displot(df['PriceEuro'])
```

```python
sns.jointplot(x=df['PriceEuro'], y=df['Area'])
```

```python
sns.jointplot(x=df['PriceEuro'], y=df['Area'], kind='hex')
```

```python
sns.jointplot(x=df['PriceEuro'], y=df['Area'], kind='kde')
```

```python
ax = sns.stripplot(x=df['Location'], y=df['PriceEuro'])
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```
